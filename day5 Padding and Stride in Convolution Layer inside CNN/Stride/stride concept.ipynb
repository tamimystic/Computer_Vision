### Stride is a parameter in a convolutional layer that determines how many pixels the filter moves across the input image at each step,, ->> default Stride 1

1. Reduces output size: By moving the filter more than one pixel at a time, a larger stride downsamples the image, making the output feature map smaller.
2. Decreases computation: A smaller output size means fewer calculations for the next layers, which makes the network faster and more computationally efficient.
3. Captures broad features: A larger stride makes the network focus on larger, more general features instead of fine details, which can help prevent overfitting.
4. Alternative to pooling: Strides are often used as a way to reduce the spatial dimensions of the input, similar to a pooling layer, but they also have learnable parameters
