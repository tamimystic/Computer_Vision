# Padding
1. Information preservation: Adding padding prevents important information at the edges of an image from getting lost, as the filter gets a chance to operate on the pixels at the borders as well.
2. Size control: Padding ensures that the size of the image does not shrink as a result of repeatedly passing it through convolutional layers.
3. Maintaining output size: Using 'same' padding allows the output image to remain the same size as the input, which helps in building deep neural networks.
4. Resolving border issues: It helps to process the pixels at the borders of the image more effectively, which increases the model's accuracy.
In essence, padding adds extra pixels around the input image, helping the model learn more effectively from the entire picture.
