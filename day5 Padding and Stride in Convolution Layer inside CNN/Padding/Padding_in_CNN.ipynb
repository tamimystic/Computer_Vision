# Padding

## With Padding

```python
# Padding

model=Sequential() #define sequential layer
model.add(Conv2D(32, kernel_size=(3,3), padding='same', activation='relu',input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(32, kernel_size=(3,3),padding='same',activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

## Without Padding

```python
withoutPaddingModel=Sequential()

withoutPaddingModel.add(Conv2D(32, kernel_size=(3,3), padding='valid', activation='relu', input_shape=(28,28,1)))
withoutPaddingModel.add(Conv2D(32, kernel_size=(3,3),padding='valid',activation='relu'))
withoutPaddingModel.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))

withoutPaddingModel.add(Flatten())

withoutPaddingModel.add(Dense(128, activation='relu'))
withoutPaddingModel.add(Dense(10, activation='softmax'))
```

```python
withoutPaddingModel.summary()
```
