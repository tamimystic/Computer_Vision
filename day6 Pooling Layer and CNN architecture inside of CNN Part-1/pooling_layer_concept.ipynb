##  WHAT: What is a Pooling Layer?
A Pooling Layer is a **down-sampling** layer in a CNN that reduces the spatial size of feature maps while keeping the most important information.  
Common window sizes: **22**, **33**  
Common stride: **2**

**Main Types of Pooling:**
- **Max Pooling**  selects the maximum value in the window  
- **Average Pooling**  takes the average value in the window

---

##  WHY: Why do we need Pooling?

1. **Reduces Computation**  
   Smaller feature maps = fewer parameters in next layers.  
   Example: 323264  (pooling)  161664

2. **Reduces Overfitting**  
   Summarizes data and removes noise.

3. **Provides Translation Invariance**  
   If a feature slightly shifts in the image, pooling still captures it.

4. **Keeps Important Features**  
   Max pooling preserves strong edges and textures.

---

##  HOW: Step-by-step Example (Max Pooling)

We have a **44 feature map**.  
We will apply **22 Max Pooling** with **stride = 2**.

### Input Feature Map (44)
```
4   2   1   5
3   6   7   2
8   4   0   3
9   1   2   6
```

### Configuration
- Window size = **2  2**  
- Stride = **2**

---

### Step-by-step Window Processing

**Step 1  Window (row 01, col 01):**
```
4  2
3  6
```
**Max = 6**

**Step 2  Window (row 01, col 23):**
```
1  5
7  2
```
**Max = 7**

**Step 3  Window (row 23, col 01):**
```
8  4
9  1
```
**Max = 9**

**Step 4  Window (row 23, col 23):**
```
0  3
2  6
```
**Max = 6**

---

### Final Output (After Max Pooling)
```
6   7
9   6
```

Explanation:  
The 44 feature map becomes **22**; each output value is the maximum from its corresponding 22 window.  
This keeps important features while reducing size.

---

##  What if we used Average Pooling?
First 22 window:
```
(4 + 2 + 3 + 6) / 4 = 3.75
```
Average pooling produces a smoother, less sharp feature map.
