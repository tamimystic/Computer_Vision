## We can download pre-trained model from keras application

```python
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
model=VGG16()
```

```python
model.summary()
```

## we can represent visually model using keras.utils plot_model

```python
from keras.utils import plot_model

plot_model(model)
```

## we can see how many model layer have here

```python
model.layers
```

## differenciate model Conv layer, and differenciate filters and biases

```python
for i in range(len(model.layers)):
  # check for convolutional layer
  if 'conv' not in model.layers[i].name:
    continue

  # get filters weight
  filters,biases=model.layers[i].get_weights()
  print("Layer number: ",i,model.layers[i].name, filters.shape)
```

```python
filters, biases=model.layers[1].get_weights()
```

```python
filters.shape
```

```python
filters
```

```python
biases.shape
```

```python
biases
```

```python
import matplotlib.pyplot as plt

n_filters=6
ix=1
fig=plt.figure(figsize=(15,10))
for i in range(n_filters):
  #get the filters
  f=filters[:,:,:,i]
  for j in range(3):
    # subplot for 6 filters and 3 channels
    plt.subplot(n_filters,3,ix)
    plt.imshow(f[:,:,j],cmap="gray")
    ix+=1

# plot the filters
plt.show()
```

## Normalize filters (0-1)

```python
f_min,f_max=filters.min(), filters.max()

filters=(filters-f_min)/(f_max-f_min)
```

```python
filters.shape
```

```python
filters
```

```python
import matplotlib.pyplot as plt

n_filters=6
ix=1
fig=plt.figure(figsize=(15,10))
for i in range(n_filters):
  #get the filters
  f=filters[:,:,:,i]
  for j in range(3):
    # subplot for 6 filters and 3 channels
    plt.subplot(n_filters,3,ix)
    plt.imshow(f[:,:,j],cmap="gray")
    ix+=1

# plot the filters
plt.show()
```

# Now apply this filters an image

```python
from tensorflow.keras import Model
```

## Create a new model from existing vgg16 model

```python
mod=Model(inputs=model.input, outputs=model.layers[1].output) #it is for first Conv layer
```

```python
mod.summary()
```

```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.utils import img_to_array
import numpy as np
from keras.applications.vgg16 import preprocess_input
```

```python
image=load_img("/content/img_p2_37.jpeg", target_size=(224,224))
image
```

```python
# convert image to array
image=img_to_array(image)
image
```

```python
image.shape
```

```python
# expand dimensions so that it represents a single 'simple'
image=np.expand_dims(image,axis=0)
image
```

## Apply filters over this image and calculating feature_map

```python
features=mod.predict(image)

fig=plt.figure(figsize=(20,25))
for i in range (1,features.shape[3]+1):
  plt.subplot(8,8,i)
  plt.imshow(features[0,:,:,i-1],cmap='gray')
```

```python
model2=VGG16()
```

```python
layer_index=[2,5,9,13,17]
outputs=[model2.layers[i].output for i in layer_index]

mod2=Model(inputs=model2.inputs, outputs=outputs)
```

```python
feature_maps_output = mod2.predict(image)

for layer_idx_val, current_fmap in zip(layer_index, feature_maps_output):
  fig = plt.figure(figsize=(20,15))
  fig.suptitle(f"Feature maps for Layer {layer_idx_val}", fontsize=20)

  num_filters_to_display = min(current_fmap.shape[3], 64) # Display up to 64 filters per layer

  rows = 8
  cols = 8

  for filter_channel_idx in range(num_filters_to_display):
    plt.subplot(rows, cols, filter_channel_idx + 1)
    plt.imshow(current_fmap[0, :, :, filter_channel_idx], cmap='gray')

  plt.show()
```
